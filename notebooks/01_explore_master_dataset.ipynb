```python
# Jupyter Notebook for initial exploration
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Load the processed data
data_path = Path("data/analysis_ready/combined_dataset.csv")
df = pd.read_csv(data_path)

print("Dataset Overview:")
print(f"Total problems evaluated: {len(df)}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst 3 rows:")
print(df.head(3))

# Basic statistics
print("\n=== Human Evaluation Statistics ===")
print(df[['correctness', 'readability', 'intention_to_use']].describe())

print("\n=== Code Metrics Statistics ===")
print(df[['code_length', 'lines_of_code', 'function_count']].describe())

# Visualizations
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Distribution plots
axes[0, 0].hist(df['correctness'], bins=5, edgecolor='black')
axes[0, 0].set_title('Distribution of Correctness Scores')
axes[0, 0].set_xlabel('Score (1-5)')

axes[0, 1].hist(df['readability'], bins=5, edgecolor='black')
axes[0, 1].set_title('Distribution of Readability Scores')
axes[0, 1].set_xlabel('Score (1-5)')

axes[0, 2].pie(df['intention_to_use'].value_counts(), 
               labels=['No', 'Yes'], autopct='%1.1f%%')
axes[0, 2].set_title('Intention to Use')

# Scatter plots
axes[1, 0].scatter(df['correctness'], df['readability'], alpha=0.6)
axes[1, 0].set_xlabel('Correctness')
axes[1, 0].set_ylabel('Readability')
axes[1, 0].set_title('Correctness vs Readability')

axes[1, 1].scatter(df['code_length'], df['correctness'], alpha=0.6)
axes[1, 1].set_xlabel('Code Length')
axes[1, 1].set_ylabel('Correctness')
axes[1, 1].set_title('Code Length vs Correctness')

# Success rate by readability
success_rate = df.groupby('readability')['intention_to_use'].mean()
axes[1, 2].bar(success_rate.index, success_rate.values)
axes[1, 2].set_xlabel('Readability Score')
axes[1, 2].set_ylabel('Intention to Use Rate')
axes[1, 2].set_title('Intention to Use by Readability')

plt.tight_layout()
plt.show()

# Save summary statistics
summary_stats = {
    'total_problems': len(df),
    'avg_correctness': df['correctness'].mean(),
    'avg_readability': df['readability'].mean(),
    'intention_to_use_rate': df['intention_to_use'].mean(),
    'high_correctness_count': len(df[df['correctness'] >= 4]),
    'high_readability_count': len(df[df['readability'] >= 4]),
}

print("\n=== Summary Statistics for Thesis ===")
for key, value in summary_stats.items():
    print(f"{key}: {value:.2f}" if isinstance(value, float) else f"{key}: {value}")
```
